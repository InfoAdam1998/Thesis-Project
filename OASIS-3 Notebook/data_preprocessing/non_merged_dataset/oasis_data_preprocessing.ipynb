* From 30-24 and CDR 0 is NC
* From 30-24 and CDR 0.5 is MCI
* From 26-20 and CDR 0.5/1 is AD

```python
import numpy as np

# Three-class mapping: 0 = NC, 1 = MCI, 2 = AD
three_class_map = {
    # NC
    "Cognitively normal":               0,
    "No dementia":                      0,

    # MCI (all “uncertain” and mild impairment labels → 1)
    "uncertain dementia":               1,
    "0.5 in memory only":               1,
    "Unc: ques. Impairment":            1,
    
    #"uncertain- possible NON AD dem":   3,

    # AD dementia and other primary dementias → 2
    "AD dem w/CVD contribut":           2,
    "AD dem w/depresss- contribut":     2,
    "AD dem w/oth (list B) contribut":  2,
    "AD dem w/PDI after AD dem contribut": 2,
    
    #"AD dem/FLD prior to AD dem":       3,
    "AD dem Language dysf prior":       2,
    "AD Dementia":                      2,
    "AD dem Language dysf after":       2,
    "AD dem w/depresss- not contribut": 2,
    "AD dem distrubed social- prior":   2,
    "AD dem w/CVD not contrib":         2,
    #"Vascular Demt- primary":           3,
    #"Frontotemporal demt. prim":        3,
    "AD dem Language dysf with":        2,
    "DAT":                              2,
    #"Dementia/PD- primary":             3,
    "AD dem w/oth (list B) not contrib":2,
    #"DLBD- primary":                    3,
    #"DLBD, primary":                    3,
    #"AD dem cannot be primary":         3,
    "AD dem w/oth unusual features/demt on": 2,
    #"Non AD dem- Other primary":        3,
}

# Apply mapping
dataset["DIAGNOSIS"] = dataset["DIAGNOSIS"].replace(three_class_map)
```

```python
dataset["DIAGNOSIS"].unique()
```

```python
values_to_remove = ['DLBD- primary', 'Vascular Demt- primary', 'Frontotemporal demt. prim',
    'Dementia/PD- primary', 'uncertain- possible NON AD dem', 'DLBD, primary', np.nan,
    'AD dem cannot be primary', '.', 'Non AD dem- Other primary', 'AD dem/FLD prior to AD dem'
]

# Drop rows where DIAGNOSIS is in the list above (including np.nan)
dataset = dataset[~dataset["DIAGNOSIS"].isin(values_to_remove)]
```
