## Correlation Analysis

Lets check the correlation between variables

Correlation measures the strength and direction of the linear relationship between two variables, but it is normalized to be independent of the units of the variables.

```python
# Import basic libraries for EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
from dotenv import load_dotenv
import os
```

```python
# Import dataset
load_dotenv()
dataset_path=os.getenv("DATASET_PATH_TWO_P")
dataset = pd.read_csv(dataset_path)
dataset.head()
```

```python
# Create a correlation dataframe
dataset_correlation = dataset.corr()
dataset_correlation
```

Now lets visualize it for better comprehension instead.

```python
# Plot a heatmap
plt.figure(figsize=(20, 20))  

sns.heatmap(dataset_correlation, 
            annot = True, 
            fmt = ".3f", 
            cmap = "coolwarm",
            square = True,
            linewidths = 0.5)
plt.title('Correlation Heatmap')
plt.show()
```

This is much better. We can clearly see which variables are correlated in a positive/negative manner. How about we set a threshold and filter only the correlated variables that are above .7 and -.7?

*Check what was the threshold in paper*

```python
plt.figure(figsize=(20, 20))  

sns.heatmap(dataset_correlation, 
            annot = True, 
            fmt = ".3f", 
            cmap = "coolwarm",
            square = True,
            linewidths = 0.5,
            mask = (dataset_correlation < 0.7) & (dataset_correlation > -0.7))
plt.title('Correlation Heatmap')
plt.show()
```

Great. Now lets create a simple pair correlation report between each variable

```python
correlation_pairs = pg.pairwise_corr(data = dataset)

correlation_report = correlation_pairs[correlation_pairs["r"].abs() > 0.7]

correlation_report[["X", "Y", "r"]].sort_values("r", ascending = False)
```

Now lets plot the positive and negative correlated variables first using scatterplots

```python
# Plot scatterplot based on the above correlation report
fig, axs = plt.subplots(5,2, figsize = (25, 20))

sns.scatterplot(data = dataset, x = "ERC SUM AVERAGE", y = "ERC CLUSTER SHADE", ax = axs[0][0])
sns.scatterplot(data = dataset, x = "ERC VARIANCE ", y = "ERC SUM VARIANCE", ax = axs[0][1])

sns.scatterplot(data = dataset, x = "ERC CORRELATION", y = "ERC SUM VARIANCE", ax = axs[1][0])
sns.scatterplot(data = dataset, x = "Hip Variance ", y = "Hip Sum Average", ax = axs[1][1])

sns.scatterplot(data = dataset, x = "Hip Correlation", y = "Hip Sum Variance", ax = axs[2][0])
sns.scatterplot(data = dataset, x = "ERC ASM", y = "ERC CLUSTER SHADE", ax = axs[2][1])

sns.scatterplot(data = dataset, x = "ERC ASM", y = "ERC ENTROPY", ax = axs[3][0])
sns.scatterplot(data = dataset, x = "ERC CONTRAST", y = "Hip Correlation", ax = axs[3][1])

sns.scatterplot(data = dataset, x = "Hip Contrast", y = "ERC SUM AVERAGE", ax = axs[4][0])
sns.scatterplot(data = dataset, x = "Hip Contrast", y = "ERC SUM AVERAGE", ax = axs[4][1])

plt.show()
```

How about by target?

```python
# Plot scatterplot based on the above correlation report
fig, axs = plt.subplots(5, 2, figsize = (25, 20))

sns.scatterplot(data=dataset, x="ERC SUM AVERAGE", y="ERC CLUSTER SHADE", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[0][0])
sns.scatterplot(data=dataset, x="ERC VARIANCE ", y="ERC SUM VARIANCE", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[0][1])

sns.scatterplot(data=dataset, x="ERC CORRELATION", y="ERC SUM VARIANCE", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[1][0])
sns.scatterplot(data=dataset, x="Hip Variance ", y="Hip Sum Average", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[1][1])

sns.scatterplot(data=dataset, x="Hip Correlation", y="Hip Sum Variance", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[2][0])
sns.scatterplot(data=dataset, x="ERC ASM", y="ERC CLUSTER SHADE", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[2][1])

sns.scatterplot(data=dataset, x="ERC ASM", y="ERC ENTROPY", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[3][0])
sns.scatterplot(data=dataset, x="ERC CONTRAST", y="Hip Correlation", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[3][1])

sns.scatterplot(data=dataset, x="Hip Contrast", y="ERC SUM AVERAGE", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[4][0])
sns.scatterplot(data=dataset, x="Hip Contrast", y="ERC SUM AVERAGE", hue="DIAGNOSIS", size="DIAGNOSIS", ax=axs[4][1])

plt.tight_layout()
plt.show()
```

This is fantastic! Isnt this more efficient instead of manually checking each column name in the pairplot? 

We can now understand how each variable is affected to one another. 

But how about the **target**? Lets narrow it down to see which features are strongly correlated with the target

```python
# Check correlation between each variable and target
correlation_with_target = dataset.corr()["DIAGNOSIS"].sort_values(ascending=False)
correlation_with_target
```

Lets plot it

```python
# Plot target correlaton using bar
plt.figure(figsize=(15,15))

correlation_with_target.plot(kind = "barh", 
                             color = "#5e76fe",
                             edgecolor = "black")
```

I believe we have a solid understanding now regarding our target feature. This will help us a lot when performing feature selection and further understand overall feature importance.
