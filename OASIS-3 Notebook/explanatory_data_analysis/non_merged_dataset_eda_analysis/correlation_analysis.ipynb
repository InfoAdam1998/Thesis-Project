## Correlation Analysis

Lets check the correlation between variables

Correlation measures the strength and direction of the linear relationship between two variables, but it is normalized to be independent of the units of the variables.

```python
# Import basic libraries for EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
from dotenv import load_dotenv
import os
```

```python
# Import dataset
load_dotenv()
dataset_path=os.getenv("DATASET_PATH_TWO_P")
dataset = pd.read_csv(dataset_path)
dataset.head()
```

```python
# Create a correlation dataframe
dataset_correlation = dataset.corr()
dataset_correlation
```

Now lets visualize it for better comprehension instead.

```python
# Plot a heatmap
plt.figure(figsize=(20, 20))  

sns.heatmap(dataset_correlation, 
            annot = True, 
            fmt = ".3f", 
            cmap = "coolwarm",
            square = True,
            linewidths = 0.5)
plt.title('Correlation Heatmap')
plt.show()
```

This is much better. We can clearly see which variables are correlated in a positive/negative manner. How about we set a threshold and filter only the correlated variables that are above .7 and -.7?

*Check what was the threshold in paper*

```python
plt.figure(figsize=(20, 20))  

sns.heatmap(dataset_correlation, 
            annot = True, 
            fmt = ".3f", 
            cmap = "coolwarm",
            square = True,
            linewidths = 0.5,
            mask = (dataset_correlation < 0.7) & (dataset_correlation > -0.7))
plt.title('Correlation Heatmap')
plt.show()
```

Great. Now lets create a simple pair correlation report between each variable

```python
correlation_pairs = pg.pairwise_corr(data = dataset, method='spearman')

correlation_report = correlation_pairs[correlation_pairs["r"].abs() > 0.7]

correlation_report[["X", "Y", "r"]].sort_values("r", ascending = False)
```

Now lets plot the positive and negative correlated variables first using scatterplots

```python
fig, axs = plt.subplots(6, 2, figsize=(25, 25))

# Remove the unused last subplot
fig.delaxes(axs[5][1])

# Plot each of the 11 correlation pairs
sns.scatterplot(data=dataset, x="ERC SUM AVERAGE", y="ERC CLUSTER SHADE", ax=axs[0][0])
axs[0][0].set_title("ERC SUM AVERAGE vs ERC CLUSTER SHADE (r = 0.95)")

sns.scatterplot(data=dataset, x="Hip Variance ", y="Hip Sum Average", ax=axs[0][1])
axs[0][1].set_title("Hip Variance vs Hip Sum Average (r = 0.82)")

sns.scatterplot(data=dataset, x="ERC VARIANCE ", y="ERC SUM VARIANCE", ax=axs[1][0])
axs[1][0].set_title("ERC VARIANCE vs ERC SUM VARIANCE (r = 0.81)")

sns.scatterplot(data=dataset, x="Hip Correlation", y="Hip Sum Variance", ax=axs[1][1])
axs[1][1].set_title("Hip Correlation vs Hip Sum Variance (r = 0.79)")

sns.scatterplot(data=dataset, x="Hip Variance ", y="Hip Sum Variance", ax=axs[2][0])
axs[2][0].set_title("Hip Variance vs Hip Sum Variance (r = 0.76)")

sns.scatterplot(data=dataset, x="ERC CORRELATION", y="ERC SUM VARIANCE", ax=axs[2][1])
axs[2][1].set_title("ERC CORRELATION vs ERC SUM VARIANCE (r = 0.72)")

sns.scatterplot(data=dataset, x="ERC ASM", y="ERC ENTROPY", ax=axs[3][0])
axs[3][0].set_title("ERC ASM vs ERC ENTROPY (r = -0.78)")

sns.scatterplot(data=dataset, x="ERC CONTRAST", y="ERC CORRELATION", ax=axs[3][1])
axs[3][1].set_title("ERC CONTRAST vs ERC CORRELATION (r = -0.84)")

sns.scatterplot(data=dataset, x="Hip Contrast", y="Hip Correlation", ax=axs[4][0])
axs[4][0].set_title("Hip Contrast vs Hip Correlation (r = -0.85)")

sns.scatterplot(data=dataset, x="ERC ASM", y="ERC CLUSTER SHADE", ax=axs[4][1])
axs[4][1].set_title("ERC ASM vs ERC CLUSTER SHADE (r = -0.87)")

sns.scatterplot(data=dataset, x="ERC ASM", y="ERC SUM AVERAGE", ax=axs[5][0])
axs[5][0].set_title("ERC ASM vs ERC SUM AVERAGE (r = -0.96)")

plt.tight_layout()
plt.show()
```

But how about the **target**? Lets narrow it down to see which features are strongly correlated with the target

```python
# Check correlation between each variable and target
correlation_with_target = dataset.corr()["DIAGNOSIS"].sort_values(ascending=False)
correlation_with_target
```

Lets plot it

```python
# Plot target correlaton using bar
plt.figure(figsize=(15,15))

correlation_with_target.plot(kind = "barh", 
                             color = "#5e76fe",
                             edgecolor = "black")
```

I believe we have a solid understanding now regarding our target feature. This will help us a lot when performing feature selection and further understand overall feature importance.
