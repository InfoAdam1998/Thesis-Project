## Correlation Analysis

Lets check the correlation between variables

Correlation measures the strength and direction of the linear relationship between two variables, but it is normalized to be independent of the units of the variables.

```python
# Import basic libraries for EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
from dotenv import load_dotenv
import os
```

```python
# Import dataset
load_dotenv()
dataset_path=os.getenv("DATASET_PATH_TWO_M")
dataset = pd.read_csv(dataset_path)
dataset.head()
```

```python
# Create a correlation dataframe
dataset_correlation = dataset.corr()
dataset_correlation
```

Now lets visualize it for better comprehension instead.

```python
# Plot a heatmap
plt.figure(figsize=(20, 20))  

sns.heatmap(dataset_correlation, 
            annot = True, 
            fmt = ".3f", 
            cmap = "coolwarm",
            square = True,
            linewidths = 0.5)
plt.title('Correlation Heatmap')
plt.show()
```

This is much better. We can clearly see which variables are correlated in a positive/negative manner. How about we set a threshold and filter only the correlated variables that are above .7 and -.7?

*Check what was the threshold in paper*

```python
plt.figure(figsize=(20, 20))  

sns.heatmap(dataset_correlation, 
            annot = True, 
            fmt = ".3f", 
            cmap = "coolwarm",
            square = True,
            linewidths = 0.5,
            mask = (dataset_correlation < 0.7) & (dataset_correlation > -0.7))
plt.title('Correlation Heatmap')
plt.show()
```

Great. Now lets create a simple pair correlation report between each variable

```python
correlation_pairs = pg.pairwise_corr(data = dataset, method='spearman')

correlation_report = correlation_pairs[correlation_pairs["r"].abs() > 0.9]

correlation_report[["X", "Y", "r"]].sort_values("r", ascending = False)
```

```python
# Set up figure with 3 rows x 2 columns (6 axes, 1 unused)
fig, axs = plt.subplots(3, 2, figsize=(25, 15))

# Plot each of the 5 correlation pairs
sns.scatterplot(data=dataset, x="Hip Variance ", y="Hip Sum Variance", ax=axs[0][0])
axs[0][0].set_title("Hip Variance vs Hip Sum Variance (r = 0.93)")

sns.scatterplot(data=dataset, x="ERC SUM AVERAGE", y="ERC CLUSTER SHADE", ax=axs[0][1])
axs[0][1].set_title("ERC SUM AVERAGE vs ERC CLUSTER SHADE (r = 0.92)")

sns.scatterplot(data=dataset, x="Hip ASM", y="Hip Sum Average", ax=axs[1][0])
axs[1][0].set_title("Hip ASM vs Hip Sum Average (r = -0.90)")

sns.scatterplot(data=dataset, x="Hip ASM", y="Hip Entropy", ax=axs[1][1])
axs[1][1].set_title("Hip ASM vs Hip Entropy (r = -0.92)")

sns.scatterplot(data=dataset, x="ERC ASM", y="ERC SUM AVERAGE", ax=axs[2][0])
axs[2][0].set_title("ERC ASM vs ERC SUM AVERAGE (r = -0.96)")

# Remove the unused subplot
fig.delaxes(axs[2][1])

# Improve layout
plt.tight_layout()
plt.show()
```

But how about the **target**? Lets narrow it down to see which features are strongly correlated with the target

```python
# Check correlation between each variable and target
correlation_with_target = dataset.corr()["DIAGNOSIS"].sort_values(ascending=False)
correlation_with_target
```

Lets plot it

```python
# Plot target correlaton using bar
plt.figure(figsize=(15,15))

correlation_with_target.plot(kind = "barh", 
                             color = "#5e76fe",
                             edgecolor = "black")
```

I believe we have a solid understanding now regarding our target feature. This will help us a lot when performing feature selection and further understand overall feature importance.
