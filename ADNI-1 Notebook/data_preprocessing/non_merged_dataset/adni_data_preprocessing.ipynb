### Data-preprocessing

```python
# Import basic libraries for EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from dotenv import load_dotenv
import os
import pingouin as pg
```

```python
# Import dataset
load_dotenv()
dataset_path=os.getenv("DATASET_PATH_R")
dataset = pd.read_excel(dataset_path)
dataset.head()
```

```python
# Keep only rows where Diagnosis is not 2
dataset = dataset[dataset["Diagnosis"] != 2].reset_index(drop=True)

# Verify
print(dataset["Diagnosis"].unique())
```

```python
# Rename class 3 to 2
dataset["Diagnosis"] = dataset["Diagnosis"].replace(3, 2)

# Verify the change
print(dataset["Diagnosis"].unique())
```

```python
dataset.drop(["RID"], axis=1, inplace=True)
```

```python
dataset["Diagnosis"].unique()
```

```python
# then shuffle rows in-place (or assign back to combined)
dataset = dataset.sample(frac=1, random_state=42).reset_index(drop=True)
```
