```python
from sklearn.model_selection import train_test_split

# Example: ADNI
adni_subjects = adni['RID'].unique()
adni_train_subs, adni_test_subs = train_test_split(
    adni_subjects, test_size=0.3, stratify=adni.groupby('RID')['Diagnosis'].first()
)

adni_train = adni[adni['RID'].isin(adni_train_subs)]
adni_test = adni[adni['RID'].isin(adni_test_subs)]

# ADNI
adni_overlap = set(adni_train['RID']).intersection(set(adni_test['RID']))
print("ADNI train-test overlap:", adni_overlap)
```

```python
oasis["DIAGNOSIS"]
```

```python
# Check datatype
print(oasis['DIAGNOSIS'].dtype)

# Check for missing values
print(oasis['DIAGNOSIS'].isnull().sum())
```

```python
# If there are missing values, fill or drop them
oasis = oasis.dropna(subset=['DIAGNOSIS'])  # simplest option

# Ensure dtype is consistent
oasis['DIAGNOSIS'] = oasis['DIAGNOSIS'].astype(str)
```

```python
oasis_subjects_df = oasis.groupby('Label')['DIAGNOSIS'].first().reset_index()
print(oasis_subjects_df['DIAGNOSIS'].value_counts())
```

```python
# Example: OASIS
oasis_subjects = oasis['Label'].unique()
oasis_train_subs, oasis_test_subs = train_test_split(
    oasis_subjects, test_size=0.3, stratify=oasis.groupby('Label')['DIAGNOSIS'].first()
)
oasis_train = oasis[oasis['Label'].isin(oasis_train_subs)]
oasis_test = oasis[oasis['Label'].isin(oasis_test_subs)]

# OASIS
oasis_overlap = set(oasis_train['Label']).intersection(set(oasis_test['Label']))
print("OASIS train-test overlap:", oasis_overlap)
```
